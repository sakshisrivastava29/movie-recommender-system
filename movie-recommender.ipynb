```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True) #where we are using this dropna fucntion to remove those cells of columns/row for which no data is available
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #here we are creating a lambda function to split the set of strings into a list 
```

```python
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] #here we are combining 5 columns into 1 column named tag to make our work easy by reducing the number of colums 
```

```python
movies.head()
```

```python
new_df = movies[['id', 'title','tags']] 
new_df
```

```python
new_df['tags']= new_df['tags'].apply(lambda x: " ".join(x)) #here we are converting tags from list to strings , split() is used to create a list whereas join() is used to make string 
new_df['tags'] 
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
#text into vector----> text vectoraization ,,here we will combine all the words in a tag and remove the stop words ,,then we wil analyise which words are occuring how much time for each movie , then compare it with other movies , and then those movies which have maxmimum smiliar words are recommended
from sklearn.feature_extraction.text import CountVectorizer
cv =  CountVectorizer(max_features = 5000,stop_words = 'english')

vector = cv.fit_transform(new_df['tags']).toarray() #here we are coverting the scipy sparsh martix into a numpy array using fittransform function 
vector.shape
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df = new_df['tags'].apply(stem)
new_df['tags']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector) #we are calculating the distance of each movie from the other one ..using the cosine_similarity function 
similarity[0] #a asimilarity matrix for 1st movie ..it show the distance of 1st moive from the rest 4806 movies
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse = True,key =lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
                         
  
```

```python
recommend('Gandhi')
```
